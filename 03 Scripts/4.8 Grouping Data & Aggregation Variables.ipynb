4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

    If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
    If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
#Testing calculation of mean for price column per user id
ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
```

```python
#Creating new column average_price 
ords_prods_merge['average_price'] = ords_prods_merge.groupby('user_id')['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
#Creating new column spending flag for high (greater and equal to 10) and low (lower 10) spenders
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

        If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
    
        If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

### Creating a test dataframe to try out procedure on small dataframe before continuing with task 7

```python
# Creating a test dataframe for user id 1
df_testuser = ords_prods_merge.loc[ords_prods_merge['user_id'] == 1]
```

```python
df_testuser
```

```python
# Identifying the median for the entire days since prior order column
df_testuser['days_since_prior_order'].median()
```

```python
# Identifying the median for days since prior order with no duplicate order numbers
df_testuser.drop_duplicates(subset = ['order_number'])['days_since_prior_order'].median()
```

Correct Median for user id 1 is 20, so I will use the drop_duplicates to proceed and identify the median for days since prior order per customer

```python
# Testing the identified example of dropping duplicates for all users
ords_prods_merge.drop_duplicates(subset = ['user_id', 'order_number']).groupby('user_id')['days_since_prior_order'].median()
```

```python
# Defining Median values as a dataframe so it can be merged in the next step
df_median_values = ords_prods_merge.drop_duplicates(subset = ['user_id', 'order_number']).groupby('user_id')['days_since_prior_order'].median()
```

```python
# Testing merge on testuser dataframe
df_testuser.merge(df_median_values, on = ['user_id'], how = 'left')
```

```python
# Performing merge on testuser dataframe
df_testuser_merge = df_testuser.merge(df_median_values, on = ['user_id'], how = 'left')
```

```python
df_testuser_merge
```

```python
# Renaming column to better match column values
df_testuser_merge.rename(columns = {'days_since_prior_order_y':'median_days_since_prior_order'})
```

### Applying results from testuser dataframe to entire dataframe

```python
# Applying test on entire dataframe
ords_prods_merge = ords_prods_merge.merge(df_median_values, on = ['user_id'], how = 'left')
```
